```python
import pandas as pd
import json
from datetime import datetime
```

# Book Reviews

```python
#loading the json file
reviews_raw = []
with open('/content/drive/Shareddrives/CIS5500-BookReviews/RawFiles/goodreads_reviews_mystery_thriller_crime.json') as file:
  for line in file:
    review_load = json.loads(line)
    reviews_raw.append(review_load)
```

```python
reviews_raw[:5]
```

```python
reviews_df = pd.DataFrame(reviews_raw)
reviews_df.head()
```

```python
#dropping unnecessary columns 
reviews_df = reviews_df.drop(columns = ['user_id','date_updated','read_at','started_at','n_comments'])
reviews_df.head()
```

```python
# Converting to date and parsing it
reviews_df['date_added'] = pd.to_datetime(reviews_df['date_added'],format ='%a %b %d %H:%M:%S %z %Y', errors = 'coerce')
reviews_df['date_added']
```

```python
#removing NaT vales
print(reviews_df[reviews_df['date_added'].isnull()])
reviews_df = reviews_df[reviews_df['date_added'].notnull()]
print(reviews_df[reviews_df['date_added'].isnull()])
len(reviews_df)
```

```python
reviews_df['date_added'].dtype
```

```python
reviews_df['year_added']= reviews_df['date_added'].apply(lambda x: x.strftime('%Y'))
reviews_df['month_added']= reviews_df['date_added'].apply(lambda x: x.strftime('%m'))
reviews_df['day_added']= reviews_df['date_added'].apply(lambda x: x.strftime('%d'))
reviews_df
```

```python
reviews_df = reviews_df.drop(columns = ['date_added'])
reviews_df
```

```python
reviews_df.to_csv(r'reviews.csv')
```

# Book Series

```python
bookseries_raw = []
with open('/content/drive/Shareddrives/CIS5500-BookReviews/RawFiles/goodreads_book_series.json') as file:
  for line in file:
    bookseries_load = json.loads(line)
    bookseries_raw.append(bookseries_load)
```

```python
bookseries_raw[:5]
```

```python
bookseries_df = pd.DataFrame(bookseries_raw)
bookseries_df.head()
```

```python
bookseries_df = bookseries_df.drop(columns = ['numbered','note'])
bookseries_df
```

```python
bookseries_df['series_works_count'] = bookseries_df['series_works_count'].astype('int32')
bookseries_df['primary_work_count'] = bookseries_df['primary_work_count'].astype('int32')
```

```python
bookseries_df.dtypes
```

```python
print(bookseries_df['series_works_count'].max())
print(bookseries_df['series_works_count'].min())
print(bookseries_df['series_works_count'].mean())
```

```python
print(bookseries_df['primary_work_count'].max())
print(bookseries_df['primary_work_count'].min())
print(bookseries_df['primary_work_count'].mean())
```

```python
bookseries_df = bookseries_df.drop(columns = ['series_works_count'])
```
